```python
import requests
```

```python
link = 'https://api.github.com'
```

Ни черта не понятно, попробуем метод .json() для повышения читабельности ответа:

```python
answ.json()
```

Уже читабельнее. Пробуем найти свой репо:

```python
user = 'VadimSpb'
```

```python
link = f'https://api.github.com/{user}/repos/'
```

```python
answ = requests.get(link)
```

```python
print(f'{answ}\n{answ.text}')
```

Упс. Читаем снова доки. Оказывается, что такая ошибка часто прилетает, когда нет запрос производится неавторизированным пользователем. Пытаемся разобраться в документации. Долго рыдаем, быстро гуглим и находим эту <a href="https://pyneng.github.io/pyneng-3/GitHub-API-JSON-example/"> статью</a>.

Для того, чтобы не светить своим паролем, импортируем библиотеку getpass. Проходим авторизацию:

```python
from getpass import getpass
```

```python
username = 'VadimSpb'
password = getpass()
```

```python
link = 'https://api.github.com/user'
```

```python
answ = requests.get(link, auth=(username, password))
```

```python
answ 
```

Аллиллуйа! Получилось. Теперь разберём ответ:

```python
answ.json()
```

Так, нам нужны репо. Отбираем методами словаря информацию по репо:

```python
link = answ.json()['repos_url']
link
```

Получаем URL на репо:

```python
answ = requests.get(link)
```

```python
answ = answ.json()
answ
```

И получаем куда больше информации, чем мы ожидали. Супер! Теперь делаем отбор имён репо:

```python
len(answ)
```

```python
for i in range(len(answ)):
    print(answ[i]['name'])
```
